```python
%matplotlib inline
import matplotlib.pyplot as plt
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar) # by default number of bins is set to 10,second array displays the bin division
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,rwidth=0.8)#proper display of bars
```

```python
plt.hist(blood_sugar,bins=5,rwidth=0.80)#setting bin size
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.80)#setting bin range
```

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.80,color='g')#setting color and histype
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('sugar range')
plt.ylabel('total number of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],rwidth=0.80,color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
#orientation for horizontal display
plt.xlabel('sugar range')
plt.ylabel('total number of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],rwidth=0.80,color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend()
```
