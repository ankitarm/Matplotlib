```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
ypos=np.arange(len(company))
ypos
plt.bar(ypos,revenue)
```

```python
ypos=np.arange(len(company))
ypos
plt.xticks(ypos,company)#for labeling
plt.bar(ypos,revenue)
```

```python
plt.ylabel("Revenue")
plt.title("US Teck Stocks")
plt.xticks(ypos,company)#for labeling
plt.bar(ypos,revenue,label="Revenue")
plt.legend()
```

```python
plt.ylabel("Revenue")
plt.title("US Teck Stocks")
plt.xticks(ypos,company)#for labeling
plt.bar(ypos,revenue,label="Revenue")
plt.bar(ypos,profit,label="Profit")
plt.legend()
```

```python
xpos=np.arange(len(company))
xpos
plt.xticks(ypos,company)
plt.ylabel("Revenue")
plt.title("US Teck Stocks")
plt.bar(xpos-0.2,revenue,label="Revenue")
plt.bar(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
xpos=np.arange(len(company))
xpos
plt.xticks(ypos,company)
plt.ylabel("Revenue")
plt.title("US Teck Stocks")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
#horizontal barchart
plt.yticks(ypos,company)
plt.xlabel("Revenue")
plt.title("US Teck Stocks")
plt.barh(ypos-0.2,revenue,label="Revenue")
plt.barh(ypos+0.2,profit,label="Profit")
plt.legend()
```
